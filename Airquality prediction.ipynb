```python
data['Datetime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'], format='%d/%m/%Y %H.%M.%S', errors='coerce')
```

```python
# FEATURE ENGINEERING
```

```python
# scale the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data[['CO(GT)']] = scaler.fit_transform(data[['CO(GT)']])
```

```python
# prepare the data for prophet

from prophet import Prophet

airquality = data[['Datetime', 'CO(GT)']].rename(columns={'Datetime': 'ds', 'CO(GT)': 'y'})
```

```python
data = data[['Datetime', 'CO(GT)']].rename(columns={'Datetime': 'ds', 'CO(GT)': 'y'})
```

```python
# splitting the data into train and test split

train = data.iloc[:-100]
test = data.iloc[-100:]
```

```python
# model training

model = Prophet()
model.fit(train)
```

```python
future = model.make_future_dataframe(periods=100, freq='H')
forecast = model.predict(future)
```

```python
# Model Evaluation
```

```python
# Extract predictions for the test set:
predictions = forecast.set_index('ds').loc[test['ds']]
```

```python
# Compute metrics:

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(test['y'], predictions['yhat'])
rmse = np.sqrt(mean_squared_error(test['y'], predictions['yhat']))
r2 = r2_score(test['y'], predictions['yhat'])

print(f"MAE: {mae}, RMSE: {rmse}, R2: {r2}")
```

```python
# Visualization:

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(test['ds'], test['y'], label='Actual')
plt.plot(test['ds'], predictions['yhat'], label='Predicted')
plt.legend()
plt.show()
```
